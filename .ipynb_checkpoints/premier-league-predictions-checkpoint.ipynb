```python
from datetime import datetime, date
  
# This function converts given date to age
def age(born):
    born = datetime.strptime(born, "%d/%m/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
  
Players['Age'] = Players['birthday'].apply(age)
  
display(Players)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Split the data into target and predictors
y = data2['team1_yellow_cards']
X = data2.drop('team1_yellow_cards', axis=1)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create the Linear Regression model object
reg = LinearRegression()

# Fit the model to the training data
reg.fit(X_train, y_train)

# Predict the target values for the test data
y_pred = reg.predict(X_test)

# Calculate the mean squared error of the model
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

# Load your data into a pandas dataframe
data = data2

# Split the data into training and test sets
train_data, test_data, train_target, test_target = train_test_split(data.drop('team1_yellow_cards', axis=1), 
                                                                      data['team1_yellow_cards'], 
                                                                      test_size=0.2)

# Initialize the model
sgd_regressor = SGDRegressor(loss='squared_loss', learning_rate='invscaling', 
                             eta0=0.001, max_iter=100)

# Fit the model to the training data
sgd_regressor.fit(train_data, train_target)

# Predict the target values for the test data
predicted_target = sgd_regressor.predict(test_data)

# Evaluate the model using mean squared error
mse = np.mean((predicted_target - test_target)**2)
print("Mean Squared Error:", mse)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split

targetVar = 'team1_yellow_cards'
corr_threshold = 0.2

def get_prediction(numerical_data, user_inputs):
    corr = spearmanr(numerical_data)
    corrSeries = pd.Series(corr[0][:,0], index=numerical_data.columns) #Series with column names and their correlation coefficients
    corrSeries = corrSeries[(corrSeries.index != targetVar) & ((corrSeries > corr_threshold)|(corrSeries < -1*corr_threshold))] #apply the threshold

    vars_to_keep = list(corrSeries.index.values) #list of variables to keep
    vars_to_keep.append(targetVar)  #add the target variable back in
    #data2 = numerical_data[vars_to_keep]
    
    # Load your data into a pandas dataframe
    cols = [ 'Age_home_0', 'Age_home_2', 'Age_home_3', 'Age_home_6', 'Age_home_9', 'Age_away_0', 'Age_away_1', 'Age_away_2', 'Age_away_3', 'Age_away_9', 'Age_away_10', 'GK_players_avg_age_away', 'MF_players_avg_age_away', 'DF_players_avg_age_away', 'FW_players_avg_age_away', '1_change', '1Manager Age', '1Longitude_birth_city', 'Latitude_stadium2', '2_change', '2Longitude_birth_city', '2d', '2pts', 'team1_yellow_cards']
    data2 = [cols]
    #data2 = data2[cols]
    
    # Split the data into training and test sets
    train_data, test_data, train_target, test_target = train_test_split(numerical_data.drop('team1_yellow_cards', axis=1), 
                                                                          numerical_data['team1_yellow_cards'], 
                                                                          test_size=0.2)
    
    # Initialize the model
    sgd_regressor = SGDRegressor(loss='squared_loss', learning_rate='invscaling', 
                                 eta0=0.001, max_iter=100)
    
    # Fit the model to the training data
    sgd_regressor.fit(train_data, train_target)
    
    # Predict the target values for the test data
    predicted_target = sgd_regressor.predict([user_inputs])
    
    return predicted_target

user_inputs = []
input_columns = list(numerical_data.columns)
input_columns.remove(targetVar)

for column in input_columns:
    user_input = float(input(f"Enter a value for {column}: "))
    user_inputs.append(user_input)

prediction = get_prediction(numerical_data, user_inputs)

print(f"The predicted target value is: {prediction[0]}")
```
